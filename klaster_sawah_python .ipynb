```python
"""
title: "K-Menas Clustering RMU dengan Python, K=5"
author: "Nasri, Dosen STIEI Malang"
date: "`7 Agustus 2022`"
"""
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt #Data Visualization 
from time import process_time
import os

t1_start = process_time()
#1. read data dan memilih kolom sebagai inpus analisa
dataset = pd.read_csv('lokasi_desa.csv')
X= dataset.iloc[:, [1,2]].values

#2. Menghitung wcss dan plot elbow, KMEANS++ Calculation, assume the max number of cluster would be 10
from sklearn.cluster import KMeans
kmeans = KMeans(init='k-means++')
wcss=[]

for i in range(1,11):
    kmeans = KMeans(n_clusters= i, init='k-means++', random_state=0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_) # wcss: Within-Cluster Sum of Square
    
#Visualizing the ELBOW method to get the optimal value of K 
plt.plot(range(1,11), wcss)
plt.title('The Elbow Method')
plt.xlabel('no of clusters')
plt.ylabel('wcss')
plt.show()

#3. Membangun Model Build, K=5 dan Menghitung koord Klaster
kmeansmodel = KMeans(n_clusters= 5, init='k-means++', random_state=0)
y_kmeans= kmeansmodel.fit_predict(X) #For unsupervised learning,y_kmeans is the final model 
df=pd.DataFrame(kmeansmodel.cluster_centers_)
print(df)

#4. Visualizing all the clusters 
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s = 10, c = 'red', label = 'Cluster 1')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s = 10, c = 'blue', label = 'Cluster 2')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], s = 10, c = 'green', label = 'Cluster 3')
plt.scatter(X[y_kmeans == 3, 0], X[y_kmeans == 3, 1], s = 10, c = 'cyan', label = 'Cluster 4')
plt.scatter(X[y_kmeans == 4, 0], X[y_kmeans == 4, 1], s = 10, c = 'magenta', label = 'Cluster 5')
#plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 50, c = 'black', label = 'Centroids')
plt.scatter(kmeansmodel.cluster_centers_[:, 0], kmeansmodel.cluster_centers_[:, 1], s = 100, c = 'black', label = 'Centroids')
plt.title('Cluster Lokasi, K=5')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.legend()
plt.show()

#5. Save output, Menghitung Waktu Eksekusi Program
df.to_csv("output_k5.csv")
df.to_excel("output_k5.xlsx")

t1_stop = process_time()
print("Waktu eksekusi program, in seconds:", t1_stop-t1_start)
```
